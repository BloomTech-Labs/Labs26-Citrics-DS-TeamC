```
import pandas as pd
```

```
pop = pd.read_csv('population.csv', encoding = "ISO-8859-1")
pop.head()
```

```
pop.columns.to_list()
```

```
cols_to_drop = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017', 'POPESTIMATE2018']
pop = pop.drop(cols_to_drop, axis=1)
pop.head()
```

```
pop = pop[pop['POPESTIMATE2019'] > 50000]
pop.shape
```

```
pop.head()
```

```
pop = pop.sort_values(by='POPESTIMATE2019', ascending=False)
pop.head()
```

```
# pop[pop['NAME']==pop['STNAME']]
pop[pop['NAME']=='District of Columbia']
```

```
pop['NAME']
```

```
for index, row in pop.iterrows():
  if (row['NAME'] == row['STNAME']) & (row['NAME'] != 'District of Columbia'):
    pop = pop.drop(index)
pop.shape
```

```
pop[pop['NAME']==pop['STNAME']]
```

```
pop['NAME'].duplicated()
```

```
pop = pop.drop_duplicates(subset=['NAME'])
pop.shape
```

```
pop[pop['NAME'].str.contains('County')]
```

```
# City = pop[pop['NAME'].str.contains('City')]
city = pop[pop['NAME'].str.contains('city')]
print(city.shape)
city.head(25)
```

```
pt = pop[pop['NAME'].str.contains('(pt.)')]
pt.shape
```

```
temp = city[~city['NAME'].isin(pt['NAME'])]
temp.shape
```

```
df = temp.copy()
```

```
df['NAME'] = df['NAME'].apply(lambda x: x[:-4])
df['NAME'] = df['NAME'].apply(lambda x: x. strip())

print(df.shape)
df.head()
```

```
df['POSTAL']= df['STNAME'].map(us_state_abbrev)
df.head()
```

```
cols = list(df)

# rearrange column order to make most sense
cols.insert(10, cols.pop(cols.index('POSTAL')))

# save new reordered frame from rearranged cols list
df = df.loc[:, cols]
df.head()
```

```
# TODO: Lower column names
col_names = df.columns.to_list()
col_names_new = [col.lower() for col in col_names]
col_names_new
```

```
df.columns = col_names_new
df.head()
```

```
df.rename(columns={'name':'city', 'stname':'state_'}, inplace=True)
df.drop(columns=['sumlev', 'state', 'county', 'place', 'cousub', 'concit', 
                 'primgeo_flag', 'funcstat'], inplace=True)
df.rename(columns={'state_':'state'}, inplace=True)
```
