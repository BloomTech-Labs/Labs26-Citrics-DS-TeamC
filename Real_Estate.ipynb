```
cities = regions[regions['region_type'] == 'city']
cities.drop(columns=['None','region_type'], inplace=True)
cities.reset_index(inplace=True)

print(cities.shape)
cities.head()
```

```
cities_spliced = cities.drop(columns=['index', 'region'])
```

```
# Split the region string into it's key parts
city_breakdown = cities['region'].str.split("; ", expand=True)
# city_breakdown.reset_index(inplace=True)
```

```
city_breakdown.head()
```

```
city_breakdown = city_breakdown.rename(columns={0:'city', 1:'state', 2:'region', 3:'county'})
# city_breakdown = city_breakdown.sort_values(['state'])
```

```
city_breakdown.head()
```

```
merged_df = pd.concat([cities_spliced, city_breakdown], axis=1)
```

```
cols = ['city', 'state', 'region', 'county']

for c in cols:
  merged_df[cols] = merged_df[cols].astype(str)
```
