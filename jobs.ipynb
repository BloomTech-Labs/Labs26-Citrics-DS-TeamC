```
final_df = pd.DataFrame(date_df)
final_df.rename(columns={0:'date'}, inplace=True)

for i in postal_dict:
  # print(i)
  postal_df = pd.DataFrame(date_df)
  postal_df.rename(columns={0:'date'}, inplace=True)

  for code in postal_dict[i]:
      data = quandl.get(f'BLSE/{code[2]}')
      data.rename(columns={'Value':code[1]}, inplace=True)
      data.reset_index(inplace=True)
      data.drop(columns=['Date'], inplace=True)
      postal_df = pd.concat([postal_df, data], axis=1, join='outer')

  postal_df['postal'] = str(i)
  final_df = pd.concat([final_df, postal_df], axis=0)

final_df.head()
```
